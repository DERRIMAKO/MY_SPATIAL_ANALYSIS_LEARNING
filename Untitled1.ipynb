```python
import pandas as pd
import geopandas as gpd
import folium
import geodatasets
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("C:/Users/USER/Downloads/archive (5)/volcano_data_2010.csv")
```

```python
df1.head()
```

## We are going to keep the relevant columns

```python
df = df1.loc[:, ["Year", "Name", "Country", "Latitude", "Longitude", "Type"]].copy()
```

```python
df.head()
```

```python
geometry = gpd.points_from_xy(df.Longitude, df.Latitude)
```

```python
geo_df = gpd.GeoDataFrame(df[["Year", "Name", "Country", "Latitude", "Longitude", "Type"]], geometry = geometry)
```

```python
geo_df.head()
```

```python
world = gpd.read_file(geodatasets.get_path("naturalearth.land"))
```

#### We have:
1. The world dataset- we are going to use as a basemap
2. geo_df- the position of the volcanoes
#### Now we Plot

```python
fig, ax = plt.subplots(figsize = (24, 18));

world.plot(
    ax=ax,
    alpha = 0.4,
    color = 'grey',
    edgecolor = 'black'
)
geo_df.plot(
    ax=ax,
    column = 'Type',
    markersize = 50,
    legend = True
)

plt.title('VOLCANOES')
```
