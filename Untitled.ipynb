```python
chicago = gpd.read_file(geodatasets.get_path('geoda.chicago_commpop'))
groceries = gpd.read_file(geodatasets.get_path('geoda.groceries'))
```

```python
chicago.plot(
    column = 'POP2010',
    cmap = 'OrRd',
    legend = True,
    legend_kwds = {
        'label':'population 2010',
        'orientation':'horizontal'
    }
)
```

### USING MATPLOLIB TO CREATE A MAP WITH LAYERS

```python
import matplotlib.pyplot as plt
```

```python
groceries = groceries.to_crs(chicago.crs)
```

```python
print(groceries.crs)
print(chicago.crs)
```

```python
fig, ax = plt.subplots(figsize = (10,10))
#ax.set_axis_off();

chicago.plot(
    ax=ax,
    color = 'none',
    edgecolor = 'black'
);

groceries.plot(
    ax=ax,
    marker = '*',
    markersize = 5,
    color = 'green'
);
#plt.show()
```
