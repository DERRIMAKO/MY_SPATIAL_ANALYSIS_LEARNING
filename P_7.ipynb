## CREATING A SIMPLE INTERACTIVE WEB MAP
### we are going to use folium

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
```

```python
m = folium.Map(location = [60.25, 24.8], tiles = 'Cartodb dark_matter', zoom_start=10, control_scale=True, prefer_canvas=True)

m
```

### ADDING LAYERS TO THE INTERACTIVE MAP

```python
m = folium.Map(location = [60.25, 24.8], tiles='CartoDB positron', zoom_start = 12, control_scale = True)

folium.Marker(
    location = [60.20426, 24.96179],
    popup = 'kampula campus',
    tooltip = 'Here you are',
    icon = folium.Icon(color='green', icon='ok-sign')
).add_to(m)

folium.LayerControl().add_to(m)

m
```
