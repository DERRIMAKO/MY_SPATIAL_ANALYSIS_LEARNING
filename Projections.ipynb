## PROJECTIONS

```python
import geopandas as gpd
import geodatasets
```

```python
#load some data
usa = gpd.read_file(geodatasets.get_path('geoda.natregimes'))
```

```python
usa.crs.name
```

```python
#Visualize
ax = usa.plot()
ax.set_title("WGS84 (lat/lon)")
```

```python
# Reproject to Albers contiguous USA
usa = usa.to_crs("ESRI:102003")
```

```python
usa.crs
```

```python
#Visualize
ax = usa.plot(figsize=(10,15))
ax.set_title("NAD 1983 Albers contiguous USA")
ax.set_axis_off()
```

```python
# Visualize
ax = usa.plot(figsize=(10,10))

ax.set_title("WGS84 (lat/lon)");

# Reproject to Albers contiguous USA
usa = usa.to_crs("ESRI:102003")

ax = usa.plot(figsize=(10,10))

ax.set_title("NAD 1983 Albers contiguous USA");
```
