# Select lands that fall under the "WILD FOREST" or "WILDERNESS" category

## Foot trails in New York state (LineString)

```python
road_trails = gpd.read_file("C:/Users/USER/Documents/Year 4/KAGGLE GEOSPATIAL ANALYSIS/DATA/DEC_roadstrails/DEC_roadstrails/Decroadstrails.shp")
```

```python
foot_trails = road_trails.loc[road_trails['ASSET']=='FOOT TRAIL'].copy()
```

## County boundaries in New York state (Polygon)

```python
counties = gpd.read_file("C:/Users/USER/Documents/Year 4/KAGGLE GEOSPATIAL ANALYSIS/DATA/NY_county_boundaries/NY_county_boundaries/NY_county_boundaries.shp")
```

# CREATING A MAP/PLOT FOR VISUALIZATION

```python
fig, ax = plt.subplots(figsize = [12, 10])
ax.set_axis_off()

counties.plot(
    ax=ax,
    color = 'none',
    edgecolor='black'
)

wild_lands.plot(
    ax=ax,
    color='grey'
)

campsites.plot(
    ax=ax,
    color = 'green',
    markersize=2
)

foot_trails.plot(
    ax=ax,
    color='red',
    markersize = 1
)
plt.show()
```
