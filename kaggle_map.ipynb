```python
# CAMPSITES IN NEW YORK STATE
POI_data = gpd.read_file("C:/Users/USER/Documents/Year 4/KAGGLE GEOSPATIAL ANALYSIS/DATA/DEC_pointsinterest/DEC_pointsinterest/Decptsofinterest.shp")
```

```python
#FOOT TRAILS IN NEW YORK STATE
road_trails = gpd.read_file("C:/Users/USER/Documents/Year 4/KAGGLE GEOSPATIAL ANALYSIS/DATA/DEC_roadstrails/DEC_roadstrails/Decroadstrails.shp")
```

```python
#COUNTY BOUNDARIES IN NEW YORK STATE
counties = gpd.read_file("C:/Users/USER/Documents/Year 4/KAGGLE GEOSPATIAL ANALYSIS/DATA/NY_county_boundaries/NY_county_boundaries/NY_county_boundaries.shp")
```

```python
counties.head()
```

### CREATING A PLOT

```python
ax = counties.plot(
    figsize = [10, 15],
    color = 'none',
    edgecolor = 'gainsboro',
    zorder = 3,
)
wild_lands.plot(
    ax=ax,
    color = 'lightgreen'
)
campsites.plot(
    ax=ax,
    markersize = 2,
    color = 'brown'
)
trails.plot(
    ax=ax,
    markersize = 1,
    color = 'black'
)
#ax.set_axis_off()
```
